##  Perceptron
ニューラルネットの基本を学ぶため、二値分類タスクをpercptronを用いて学習する。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Perceptron(object):
    
    def __init__(self, eta=0.001, n_iter=1000):
        self.eta = eta # 学習率
        self.n_iter = n_iter # 繰り返し回数
        
    def fit(self, X, y):
        self.w = np.zeros(1+X.shape[1])
        self.errors = []
        
        for _ in range(self.n_iter):
            error = 0
            for x, target in zip(X,y):
                dif = target - self.predict(x)
                print(dif)
                update = dif * self.eta
                self.w[1:] += update * x
                self.w[0] += update
                error += int(update != 0.0)
            self.errors.append(error)
        return self
    
    def net_input(self, X):
        return np.dot(X, self.w[1:] + self.w[0])
    
    def predict(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, 0)
```

```python
X = np.array([[0,0],[1,0],[0,1],[1,1], [0,1.5]])
y = np.array([1,1,1,0,1])
colors = ["r", "b"]
for p, q in zip(X,y):
    plt.plot(p[0],p[1],"o", c=colors[q])
    plt.xlim([-1,2])
    plt.ylim([-1,2])
```

```python
pct = Perceptron()
pct.fit(X,y)
```

```python
_x = np.arange(-1,2,0.1)
_w = pct.w
_y = [(_w[0]+_w[1]*x)/(-_w[2]) for x in _x]
plt.plot(_x, _y)
for p, q in zip(X,y):
    plt.plot(p[0],p[1],"o", c=colors[q])
    plt.xlim([-1,2])
    plt.ylim([-1,2])
```
